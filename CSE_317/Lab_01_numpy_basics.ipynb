### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(start=-1.5, stop= 2 * np.pi, step= 0.001)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(start=-1.5, stop= 2 * np.pi, step=0.001)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('Steps')
plt.ylabel('Values')
plt.title('Sine and Cosine Graph')
plt.legend(['Sine','Cosine'], loc='lower right')
plt.show()
```

```python
x = np.arange(start= -1.5, stop= 2 * np.pi, step= 0.001)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine Curve')
plt.show()

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine Curve')
plt.show()
```

```python
x = np.arange(-3, 1* np.pi, 0.01)
y = np.arange(-3, 1* np.pi, 0.01)

x,y = np.meshgrid(x,y)
z = x*np.exp(-x**2-y**2)+25

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x,y,z,cmap='viridis', edgecolor='none')
ax.set_title('Random Plot')
plt.show()                
```
